```python
# Bar Graph of Top 10 Artist with Most Songs
```

```python
def top_10_artists_bar_plot(Top_10_Artist):
    """
    This fucntion draws a bar chart to show the top 10 artists who has highest number of songs in the year 2023
    
    """
    plt.figure(figsize=(10, 6))
    barplot = sns.barplot(x=Top_10_Artist.index, y=Top_10_Artist.values, palette='viridis')
    plt.title('Top 10 Artists with Most Top Songs on Spotify')
    plt.xlabel('Artists Name')
    plt.ylabel('Number of Songs')
    plt.ylim(0, 40)
    plt.xticks(rotation=45)

    for bars in barplot.patches:
        barplot.annotate(format(bars.get_height(), '.0f'), 
                         (bars.get_x() + bars.get_width() / 2., bars.get_height()), 
                         ha='center', va='center', 
                         xytext=(0, 10), 
                         textcoords='offset points')
    plt.tight_layout()
    plt.savefig('top_10_artists_2023_bar_graph.png', dpi=300)

    plt.show()

top_10_artists_bar_plot(Top_10_Artist)
```

```python
# Scatter Plot
```

```python

def valence_vs_energy_scatter_plot(New_file):
    """
    Draws a scatter Graph comparing 'Valence' and 'Energy' for a given dataset.

    This function takes a DataFrame[New_file] as input and creates a scatter plot to visualize
    the relationship between 'Valence' and 'Energy' percentages of songs. 

    
    """
    plt.figure(figsize=(8, 6))
    plt.scatter(New_file['Valence'], New_file['Energy'], color='green', alpha=0.5, edgecolors='w', s=80)
    plt.title("Valence vs Energy ")
    plt.xlabel('Valence')
    plt.ylabel('Energy')
    plt.grid(True)
    plt.savefig('top_10_artists_2023_scatter_graph.png', dpi=300)
    plt.show()

valence_vs_energy_scatter_plot(New_file)


    
```

```python
# Correlation Heat map Analysis
```

```python
def correlation_heatmap_plot(New_file):
    """
    Creates a correlation heatmap for numerical columns in the given DataSet.

    This function selects numerical columns from the provided DataFrame[New_file] and calculates
    the correlation matrix using seaborn's heatmap function.

    """
    numeric_file = New_file.select_dtypes(include=['float64', 'int64'])

    Correlation_Matrix = numeric_file.corr()
    plt.figure(figsize=(14, 12))
    sns.heatmap(Correlation_Matrix, annot=True, cmap='coolwarm')
    plt.title('Correlation Heatmap')
    plt.savefig('top_10_artists_2023_heat_map.png', dpi=500)
    plt.show()
correlation_heatmap_plot(New_file)
```
