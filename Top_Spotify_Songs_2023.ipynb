```python
# Relationship Analysis
```

```python
Artist_Count = New_file['Artist(s)_Name'].value_counts()
Artist_Count
```

```python
Top_10_Artist = Artist_Count.head(10)
```

```python
colors = ['red', 'blue', 'green', 'orange', 'purple', 'pink', 'yellow', 'gray', 'violet', 'black']
Top_10_Artist.plot(kind='bar',color = colors)
plt.title('Top 10 Artists Most Songs in 2023')
plt.xlabel('Artist(s)_Name')
plt.ylabel('Artist_Count')
plt.tight_layout()
plt.show()
```

```python
x = range(len(New_file))
plt.scatter(x,New_file['Valence_%'], color='blue', label='Valence')
plt.scatter(x,New_file['Energy_%'], color='red', label='Energy')
plt.title('Valence_%   vs  Energy_%')
plt.xlabel('Valence')
plt.ylabel('Energy')
plt.show()
```

```python
numeric_file = New_file.select_dtypes(include=['float64', 'int64'])

Correlation_Matrix = numeric_file.corr()
plt.figure(figsize=(14, 12))
sns.heatmap(Correlation_Matrix,annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
